```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('../../')  # 'class_func.py' dosyası üst klasörde olduğu için onu path'e eklendi

import class_func as cf # 'class_func.py' dosyasını import edildi
import numpy as np # numpy kütüphanesini import edildi
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('minima_pri_final.txt', delimiter='\t', comment="#", names=['BJD', 'p/s', 'type', 'error', 'ref', "1", "2", "3"])
print(df.columns)
```

```python
mintimes = df["BJD"].values # veriden minimum zamanlar alındı
mintypes = (df["p/s"] - 1).values # veriden minimum zamanların türleri aldındı. Bu türler 1. minimumlar için 0 2. minimumlar için 1 olacak şekilde
errors = df["error"].values # veriden hata değerleri alındı
ref_mintime = 2453174.442769 # minimum zamanın referans zamanı
ref_period = 0.101015969 # periyot referans zamanı
data = cf.OC_data(Mintimes=mintimes, Mintypes=mintypes, Ref_mintime=ref_mintime, Ref_period=ref_period) # Eldeki verilerle OC_data nesnesi oluşturuldu
```

```python
data.plot_OC() # Bu veriler isteğe bağlı olarak plot_OC fonksiyonu ile grafiğe dönüştürülebilir.
```

```python
LTE1_params = {"amp":7.57e-05, "P_LiTE":cf.Parameter(3195, "day"), "T_LiTE":cf.Parameter(2453619, "BJD"), "e":0, "omega":cf.Parameter(0, "deg")}
LTE1 = cf.LiTE(params=LTE1_params, name="LTE1")
LTE2_params = {"amp":0.000254, "P_LiTE":cf.Parameter(8404, "day"), "T_LiTE":cf.Parameter(2451001, "BJD"), "e":0, "omega":cf.Parameter(0, "deg")}
LTE2 = cf.LiTE(params=LTE2_params, name="LTE2")

epochs = np.linspace(-40000, 70000, 1000)
model = cf.OC_model(epochs=epochs, Ref_period=ref_period, Ref_mintime=ref_mintime)
model.add_model_component(LTE1) # OC nesnesine LiTE1 adlı bileşen eklendi
model.add_model_component(LTE2) # OC nesnesine LiTE2 adlı bileşen eklendi
model.summary()

# For residuals
model_ = cf.OC_model(epochs=data.Ecorr, Ref_period=ref_period, Ref_mintime=ref_mintime)
model_.add_model_component(LTE1) # OC nesnesine LiTE1 adlı bileşen eklendi
model_.add_model_component(LTE2) # OC nesnesine LiTE2 adlı bileşen eklendi
```

```python
plt.plot(data.Ecorr, data.OC, "b.")
plt.plot(model.epochs, model.calculate_oc(), "r-")
```

```python
plt.plot(data.Ecorr, data.OC-model_.calculate_oc(), "b.")
plt.axhline(0,color="k",alpha=0.5,ls=":")
```
